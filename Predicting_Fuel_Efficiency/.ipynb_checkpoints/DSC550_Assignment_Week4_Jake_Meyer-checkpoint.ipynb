### Part 2 - Begin by prepping the data for modeling.

```python
'''
Remove the car name column from the data frame.
'''
df.drop(columns = "car name", inplace = True)
df.head()
```

The 'horsepower' column contains the ? symbol which is why the column is showing up as a string/object data type.

```python
'''
Replace the ? with the column mean.
Start by converting the ? value to NaN.
'''
df['horsepower'] = df['horsepower'].replace('?', np.nan)
```

```python
'''
Convert the horsepower column to a float data type.
'''
df['horsepower'] = pd.to_numeric(df['horsepower'])
```

```python
'''
Replace the NaN values with the calculated mean for the horsepower column.
Index Number 374 was NaN and should show the mean calculation replaced as evidence.
'''
mean_calc = df['horsepower'].mean()
df['horsepower'] = round(df['horsepower'].replace(np.nan, mean_calc),1)
df.tail(25)
```

```python
'''
Confirm the data type for the columns after the change to horsepower.
'''
df.dtypes
```

```python
'''
Understand the categories and counts for the origin column.
'''
df['origin'].value_counts()
```

```python
'''
Create dummy variables for the origin column.
View the data frame (df) to ensure the dummy variable columns were created.
'''
df[['origin_1', 'origin_2', 'origin_3']] = pd.get_dummies(df['origin'])
df.head()
```

```python
'''
Verify the dummy variables match the count recently found from the value_counts() output above.
'''
print('Origin 1 Sum: {}\n'\
      'Origin 2 Sum: {}\n'\
      'Origin 3 Sum: {}'.format(df['origin_1'].sum(), df['origin_2'].sum(), df['origin_3'].sum()))
```

```python
'''
Remove the origin column from the data frame. The dummy variables were created to represent
the three categories.
'''
df.drop(columns = "origin", inplace = True)
df.head()
```

### Part 5 - Randomly split the data into 80% training data and 20% test data, where your target is mpg.

```python
'''
Define the target variable as y (mpg) and independent variables as X. 
'''
x = df.drop('mpg', axis = 1)
y = df['mpg']
```

```python
'''
Show the x data frame for the independent variables cylinders, displacement, horsepower,
weight, acceleration, model year, origin_1, origin_2, origin_3.
'''
x.head()
```

```python
'''
Show the y data frame for the target variable mpg.
'''
y.head()
```

```python
'''
Split the data with train_test_split from sklearn.
Use test_size = 0.2 to split the data into 80% training and 20% testing data.
'''
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
print("The x_train shape is {} rows and {} columns.".format(x_train.shape[0],x_train.shape[1]))
```

```python
print("The y_train shape is {} rows.".format(y_train.shape[0]))
```

```python
print("The x_test shape is {} rows and {} columns.".format(x_test.shape[0],x_test.shape[1]))
```

```python
print("The y_test shape is {} rows.".format(y_test.shape[0]))
```

###  Part 6 - Train an ordinary linear regression on the training data.

```python
'''
Create the linear regression model with the training data.
'''
regressor = linear_model.LinearRegression()
model = regressor.fit(x_train, y_train)
model
```

```python
'''
View the intercept of the model.
'''
model.intercept_
```

```python
'''
View the coefficients of the model.
'''
model.coef_
```

```python
'''
Create predictions for mpg based on training data set.
'''
y_predict_trained = regressor.predict(x_train)
y_predict_trained
```

```python
'''
Create predictions for mpg based on test data set.
'''
y_predict_test = regressor.predict(x_test)
y_predict_test
```

### Part 7 - Calculate R-Squared, RMSE, and MAE on both the training and test sets and interpret your results.

```python
'''
Calcualte the R-Squared value for the model based on the training data set.
'''
R2_train = r2_score(y_train, y_predict_trained)
print('R-Squared value for model based on training data set is {}.'.format(R2_train))
```

```python
'''
Calculate the adusted R-Squared value for the model based on the training data set.
'''
n = x_train.shape[0]
p = x_train.shape[1]

adj_R2_train = 1 - (1 - R2_train) * ((n - 1)/(n-p-1))

print('Adjusted R-Squared value is {}'.format(adj_R2_train))
```

```python
'''
Calculate the Root Mean Squared Error for the model based on the training data set.
'''
mse_train = mean_squared_error(y_train, y_predict_trained)
rmse_train = np.sqrt(mse_train)
print('Root Mean Squared Error for model based on training data set is {}'.format(rmse_train))
```

```python
'''
Calculate the Mean Absolute Error for the model based on the training data set.
'''
mae_train = mean_absolute_error(y_train, y_predict_trained)
print('Mean Absolute Error for model based on training data set is {}'.format(mae_train))
```

```python
'''
Calcualte the R-Squared value for the model based on the test data set.
'''
R2_test = r2_score(y_test, y_predict_test)
print('R-Squared value for model based on test data set is {}.'.format(R2_test))
```

```python
'''
Calculate the adusted R-Squared value for the model based on the test data set.
'''
n = x_test.shape[0]
p = x_test.shape[1]

adj_R2_test = 1 - (1 - R2_test) * ((n - 1)/(n-p-1))

print('Adjusted R-Squared value is {}'.format(adj_R2_test))
```

```python
'''
Calculate the Root Mean Squared Error for the model based on the test data set.
'''
mse_test = mean_squared_error(y_test, y_predict_test)
rmse_test = np.sqrt(mse_test)
print('Root Mean Squared Error for model based on test data set is {}'.format(rmse_test))
```

```python
'''
Calculate the Mean Absolute Error for the model based on the test data set.
'''
mae_test = mean_absolute_error(y_test, y_predict_test)
print('Mean Absolute Error for model based on test data set is {}'.format(mae_test))
```

Prior to interpreting the results for R-Squared, Root Mean Squared Error (RMSE), and Mean Absolute Error (MAE), it must be understood what each of these terms represents. The R-Squared value measures how much variance in the target feature (dependent variable) is explained within the model. The R-Squared value will help signify how strong of a linear relationship exists to the fitted regression line by taking the explained variation divided by the total variation. The adjusted R-Squared value will take the number of independent variables in the model into condsideration. The RMSE value represents the standard deviation of the residuals and is calculated by taking the square root of the Mean Squared Error (MSE). The MSE value measures the variance of the residuals and is calculated by taking the average squared difference between the predicted and original values. The MAE value essentially measures the average (absolute) difference of the residuals. 

For the training data set, the R-Squared and adjusted R-Squared values of the model were 0.82 and 0.81 respectively. For the test data set, the R-Squared and adjusted R-Squared values of the model were 0.85 and 0.82 respectively. The model does a pretty decent job for explaining the variance for mpg on both training and test sets of data, however the R-Squared and adjusted R-Squared values are higher for the model when applied on the test data. The RMSE for the model with the trained data set was 3.37. The RMSE for the model with the test data was 2.89. The standard deviation of the residuals was slightly lower for the model when applied to the test data rather than the training data. The MAE for the model based on the training data was 2.61 whereas the MAE for the model based on test data was 2.29. The regression line fits better for the test data compared to the training data.
